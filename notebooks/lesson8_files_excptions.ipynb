## Reading and Writing Files
- To read a file, you can use the `open()` function to create a file object, and then call the `read()` method to read the contents of the file. For example:
```python
with open('filename.txt', 'r') as file:
    contents = file.read()
    print(contents)
```
- To write to a file, you can use the `open()` function with the 'w' mode, and then call the `write()` method to write data to the file. For example:
```python
with open('filename.txt', 'w') as file:
    file.write('Hello, World!')
```
- You can also use the 'a' mode to append data to an existing file without overwriting it. For example:
```pythonwith open('filename.txt', 'a') as file:
    file.write('This is an additional line.')
```
## Handling Exceptions
- Exceptions are errors that occur during the execution of a program. You can handle exceptions using `try` and `except` blocks. For example:
```pythontry:
    result = 10 / 0  

except ZeroDivisionError:
    print("You cannot divide by zero!")
```- You can also catch multiple exceptions by specifying them in a tuple. For example:
```pythontry:
    value = int(input("Enter a number: ")) 
except (ValueError, TypeError):
    print("Invalid input! Please enter a valid number.")
```- You can use the `else` block to execute code that should run if no exceptions were raised. For example:
```pythontry:
    value = int(input("Enter a number: "))
except ValueError:
    print("Invalid input! Please enter a valid number.")
else:
    print(f"You entered: {value}")```- Finally, you can use the `finally` block to execute code that should run regardless of whether an exception was raised or not. For example:
```pythontry:
    value = int(input("Enter a number: "))
except ValueError:
    print("Invalid input! Please enter a valid number.")
finally:    print("This will always be executed.")```       

## Conclusion
In this lesson, we covered how to read and write files in Python, as well as how to handle exceptions using `try`, `except`, `else`, and `finally` blocks. These concepts are essential for working with files and managing errors in your Python programs.
