# Naming rule of thumb (good habits matters)
 It can take some practice to learn how to create good variable names, especially as your programs become more interesting and complicated. As you write more programs and start to read through other people’s code, you’ll get better at coming up with meaningful names.
1. Variable names should be short but descriptive. 
For example, name is better than n, student_name is better than s_n.
2. Be careful when using the lowercase letter l and the uppercase letter O because they could be confused with the numbers 1 and 0.
3. The Python variables at this point should be lowercase. You won’t get errors if you use uppercase letters, but it’s a good idea to avoid using them for now.

# 1 Strings
- f-Strings and some string operations .title() .upper() .lower()
- "add a \n new line" 
- "add a \ttab"
- Stripping Whitespace: For example, one important instance might involve checking people’s usernames when they log in to a website. Extra whitespace can be confusing in much simpler situations as well. Python can look for extra whitespace on the right and left sides of a string. To ensure that no whitespace exists at the right end of a string, use the .rstrip() .lstrip() .strip() method which removes spaces \n new lines and \t tabs. 
- data types in Python: str, int, float
    to convert int and float into str: str()


```python
name = "i love you"
print(name.title()) # Changing Case in a String with Methods
print(name.upper()) # Changing Case in a String with Methods
print(name.lower()) # Changing Case in a String with Methods
first_name = "python"
last_name = "green"
full_name = f"{first_name} {last_name}"
print(first_name + " " + last_name)
print(full_name) # Using f-Strings to Combine Strings
print(f"Hello, {full_name.title()}!") # Using f-Strings to Combine Strings
message = f"Hello, {full_name.title()}!"
print(message) # Using f-Strings to Combine Strings
```

```python
favourite_language = " python      "
print(f"Hej, {favourite_language.title()} is my favourite language!") # Stripping Whitespace
print(f"Hej, {favourite_language.rstrip().title()} is my favourite language! with .rstrip()") # Stripping Whitespace
print(f"Hej, {favourite_language.lstrip().title()} is my favourite language! with .lstrip()") # Stripping Whitespace
print(f"Hej, {favourite_language.strip().title()} is my favourite language! with .strip()") # Stripping Whitespace
print(f"Hej, {favourite_language.title()} is my favourite language! \n\tBut .rstrip() does not remove the trailing whitespace permanently.") # Stripping Whitespace
```
